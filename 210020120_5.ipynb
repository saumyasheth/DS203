```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading the Data set and printing it

```python
df = pd.read_csv('NDAP_REPORT_7004.csv') #reading csv file into a pandas dataframe
display(df)  #displaying the dataframe to check it
```

### Cleaning Data

```python
df=df.fillna(0) #filling the nan values with 0
df['Number of census houses'] = df['Number of census houses'].replace({'zero' : 0}) #replacing string zero with 0 in this column
df['Houses having grass or thatch or bamboo etc. as material of wall'] = df['Houses having grass or thatch or bamboo etc. as material of wall'].replace({'zero' : 0}) #replacing string zero with 0 in this column
df['Houses having plastic or polythene as material of wall'] = df['Houses having plastic or polythene as material of wall'].replace({'zero' : 0}) #replacing string zero with 0 in this column
```

```python
#changing data type of three columns to float
df[["Number of census houses", "Houses having grass or thatch or bamboo etc. as material of wall", "Houses having plastic or polythene as material of wall"]] = df[["Number of census houses", "Houses having grass or thatch or bamboo etc. as material of wall", "Houses having plastic or polythene as material of wall"]].apply(pd.to_numeric)
```

### Q 1. a) Q-Q plot for gaussian and uniform distribution for the given column and checking which one is a better fit

```python
import statsmodels.api as sm
import scipy.stats as stats
fig = sm.qqplot(df.iloc[:,21] , dist=stats.norm) #qq plot for normal distribution
plt.title(df.columns[21])
fig1 = sm.qqplot(df.iloc[:,21] , dist=stats.uniform) #qqplot for uniform distribution
plt.title(df.columns[21])
plt.show()    
```

```python
print('Looking at both the plots,Q-Q plot for Gaussian distribution is a better fit') #because in between slope in plot for gaussian is more than the slope of uniform and if graph is more nearer to 45 degree straight line, it explains better
```

### Q 1 d)  (calculation also present in pdf uploaded)

```python
from numpy import log as ln
import math
b = df.iloc[:,21].max() #finding the maximum value of column
print(b)
a = df.iloc[:,21].min() #finding the minimum value of column
print(a)
variance = df.iloc[:,21].var() #finding the variance of column
print(variance)
stdev = pow(variance,0.5) #finding standard deviation byy finding square root of variance
print(stdev)
meanofcolumn = df.iloc[:,21].mean() #finding the mean of column
print(meanofcolumn)
loglikelihoodnormal = (((1473*(ln(2*math.pi*variance)))/2) + (1473/2))*(-1) #finding log likelihood for normal distribution, derivation of expression done in previous question
print(loglikelihoodnormal)
loglikelihooduniform = (1473*ln(b-a))*(-1) #finding log likelihood for uniform distribution, derivation of expression done in previous question
print(loglikelihooduniform)
if(loglikelihoodnormal>loglikelihooduniform): #the one which has higher log likelihood is more likely to explain the data
    print("Gaussian with MLE parameters is more likely to explain the observed data")
else:
    print("Uniform with MLE parameters is more likely to explain the observed data")    
```

### Q3 c)   (the result of chi-square test and its reason is mentioned in the end)

```python
#print(automobile_df['body-style'].unique()) #did this to check how many filters need to be applied
print('The test used here is Chi-square test. This test is used because we have two categorical data and this is a good method to check if there is a relation between two data having categorical values')
# Below this are all the steps of chi-square test
filter1 = automobile_df['body-style'] == 'convertible'
filter2 = automobile_df['body-style'] == 'hatchback'
filter3 = automobile_df['body-style'] == 'sedan'
filter4 = automobile_df['body-style'] == 'wagon'
filter5 = automobile_df['body-style'] == 'hardtop'
data = {'convertible' : automobile_df[filter1]['drive-wheels'].value_counts() , 'hatchback' : automobile_df[filter2]['drive-wheels'].value_counts(), 'sedan' : automobile_df[filter3]['drive-wheels'].value_counts(), 'wagon' : automobile_df[filter4]['drive-wheels'].value_counts(), 'hardtop' : automobile_df[filter5]['drive-wheels'].value_counts()}
new_df = pd.DataFrame(data)
new_df = new_df.fillna(0)
print(new_df)
```

```python
observed_df = new_df.transpose()
print(observed_df)
```

```python
#expected_df = pd.DataFrame()
total = 205
row_total = np.empty(5, dtype=float) 
row_total[0] = 6
row_total[1] = 70
row_total[2] = 96
row_total[3] = 25
row_total[4] = 8
column_total = np.empty(3, dtype=float)
column_total[0] = 9
column_total[1] = 120
column_total[2] = 76
data = {'4wd' : [6*9/205,70*9/205,96*9/205,25*9/205,8*9/205], 'fwd' : [6*120/205,70*120/205,96*120/205,25*120/205,8*120/205], 'rwd' : [6*76/205,70*76/205,96*76/205,25*76/205,8*76/205]}
expected_df = pd.DataFrame(data , index = ['convertible','hatchback','sedan','wagon','hardtop'])
print(expected_df)
```

```python
p_df = pd.DataFrame()
p_df = (pow((expected_df - observed_df),2))/expected_df
print(p_df)
p = p_df.sum().sum()
print('Value of p is :')
print(p)
```

```python
doffreedom = (3-1)*(5-1)
print('Value of p is : '+str(p))
print('Degree of freedom is : '+str(doffreedom))
print('v is the degree of freedom which is 8 in this case')
print(' ν           0.90      0.95     0.975      0.99     0.999')
print(' 8          13.362    15.507   17.535     20.090   26.125')
print('for alpha = 0.001 that is for 0.001 error the critical value for chi square distribution is 26.125 as it can be seen in the table')
print('The value of p comes out to be greater than the critical value, therefore we reject the null hypothesis')
print('There is a relation between the columns body-style and drive-wheels')
```

### The result of the chi-square test is that there is a relation between the body-style and drive wheels
